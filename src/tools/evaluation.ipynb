```python
import os
import base64
import json
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['CVLab[2]', 'CrossMatch1', 'CrossMatch2', 'CrossMatch3']
threshold_36 = [86.1, 86.58, 80.48, 45.12]
threshold_48 = [75.6, 70.73, 70.73, 28.04]
threshold_60 = [61.5, 60.97, 58.53, 10.97]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, threshold_36, 0.22, label='36 = 0.1')
rects2 = ax.bar(x, threshold_48, 0.22, label='48 = 0.01')
rects3 = ax.bar(x + width, threshold_60, 0.22, label='60 = 0.001')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage')
ax.set_title('Comparision by Results')
ax.set_xticks(x, labels)
ax.legend(title='FAR(%)')

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.set_yticks(np.arange(0, 110, 10))

fig.set_figwidth(9)
fig.tight_layout()

plt.show()
```
